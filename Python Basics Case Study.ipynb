# **1. Grocery Store Bill Calculator**

A grocery store wants a program to calculate the total bill for a customer.


1.   Store the prices of items in a dictionary (e.g., {'apple': 10, 'milk': 50})
2.   Take the items purchased and their quantities as input.
1.   Calculate and display the total bill amount.




```python

print("Welcome to the Grocery Store!")

#storing the prices of items in a dictionary
products = { 'apple': 10, 'milk': 50, 'eggs': 10, 'coffee': 25, 'biscuits': 30, 'juice': 15, 'ice cream': 35, 'cakes': 100, 'broom': 200, 'strawberry': 95 }

#initialize the total bill amount
total_bill = 0

#loop to take input for multiple items
while True:
    purchased_item = input("Enter the item purchased or type done when finished : ")

    #if the user wants to finish shopping
    if purchased_item == 'done':
        break

    #if the item is available
    if purchased_item not in products:
        print("Sorry, the item is unavailable.")
        continue

    #quantity of item purchased
    quantity = int(input(f"Enter the quantity of {purchased_item}: "))
    if quantity < 0:
      print("Quantity cannot be less than zero.")
      continue

    #calculating the cost of purchsed items

    price = products[purchased_item]
    total_bill += price * quantity

    print(f"{quantity} of {purchased_item} is added to the cart. Your total: {total_bill}")

#final bill
print(f"Your total bill is: {total_bill}")
print("Thank you for shopping with us. Goodbye!")
```

# 2. Student Grade Calculator

A teacher wants a program to calculate grades for students based on their marks.


1.    Store marks for 5 subjects in a list.
2.   Calculate the average and determine the grade:
1.  Display the average and the grade.

*  90+ -> A
*  70-89 -> B
*  50-69 -> C
*  <50 -> F



```python
#storing marks for five subjects in a list
print("Enter marks for 5 subjects (out of 100):")
marks = []
for i in range(1, 6):  #5 subjects
    while True:
            mark = int(input(f"Subject {i}: "))
            if 0 <= mark <= 100:
                marks.append(mark)
                break
            else:
                print("The marks must be between 0 and 100. Please try again.")


#the average mark
average_mark = sum(marks) / len(marks)

#the grade based on the average
if average_mark >= 90:
    grade = 'A'
elif 70 <= average_mark < 90:
    grade = 'B'
elif 50 <= average_mark < 70:
    grade = 'C'
else:
    grade = 'F'

#displaying the average marks and grade
print(f'The Average Mark is {average_mark:.2f} and the Grade is {grade}')
```

# 3. A movie theatre wants to automate ticket booking for customers.



1. Use a list to store the available seats (e.g., [1, 2, 3, ..., 20]).
2.  Allow the customer to book a seat by entering the seat number.
1.   After booking, remove the seat from the list of available seats.
2.   Display the remaining available seats.





```python
print("Welcome to the Movie Theatre Ticket Booking System!")
#list to store the available seats
available_seats = list(range(1, 21))  #seats from 1 to 20


print(f"Available seats: {available_seats}")

while True:
    #displaying the available seats
    print(f"Current available seats: {available_seats}")

    #asking the user to select a seat
    try:
        seat_number = int(input("Enter the seat number you want to book or type 0 to exit : "))

        if seat_number == 0:
            print("Thank you for visiting. Goodbye!")
            break

        if seat_number not in available_seats:
            print("Sorry, that seat is either already booked or invalid. Please choose another seat.")
            continue

        #booking the seat (removing it from the list)
        available_seats.remove(seat_number)
        print(f"Seat {seat_number} has been successfully booked!")

    except ValueError:
        print("Invalid input. Please enter a valid seat number.")
```
