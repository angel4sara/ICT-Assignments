## 5.  Check and handle outliers in at least 3 columns in the dataset

```python
sns.boxplot(titanic['Age'], color = 'lightblue')
```

```python
Q1 = np.percentile(titanic['Age'],25)
Q3 = np.percentile(titanic['Age'],75)
IQR = Q3 -Q1
lower_limit = Q1 - (1.5 * IQR)
upper_limit = Q3 + (1.5 * IQR)
print(f"The 25th quartile is : {Q1}")
print(f"The 75th quartile is : {Q3}")
print(f"The IQR is : {IQR}")
print(f"The lower limit is : {lower_limit}")
print(f"The upper limit is : {upper_limit}")
```

```python
outlier = []
for x in titanic['Age']:
  if x<lower_limit or x>upper_limit:
    outlier.append(x)
print(f"The Outliers are : {outlier}")

ind = titanic['Age'] > upper_limit
titanic.loc[ind].index
```

**We could separately handle the outliers as we did in age or we could do them together like below.**

```python
# Identify and handle outliers using IQR method
columns_to_check = [ 'Fare', 'SibSp']

for column in columns_to_check:
    Q1 = titanic[column].quantile(0.25)
    Q3 = titanic[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Capping the outliers to the bounds
    titanic[column] = titanic[column].clip(lower=lower_bound, upper=upper_bound)


titanic[columns_to_check].describe()
```
