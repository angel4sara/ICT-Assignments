## Visualisations

###  5. Perform meaningful visualizations using the dataset. Bring at least 3 meaningful visualizations

### **Heatmap**

```python
corr = data_iris.iloc[:, :-1].corr()
sns.heatmap(corr, annot =True, cmap ='RdBu')
plt.title('Correlation Matrix')
plt.show()
```

From the Heatmap above:
* Petal Length and Petal Width has the highest positive correlation followed by Petal Length and Sepal Length.
*Petal Length and Sepal Width has highest negative correlation followed by Petal Width and Sepal Width.

### **Pair Plot**

```python
Plots = sns.pairplot(data_iris, hue='Classification')
Plots.fig.suptitle('Pairplot of the Iris Data', y=1.02)
plt.show()
```

The pairplot depicts:
* The Iris-setosa has a clear distinction or separation from other groups in terms of Sepal Length, Sepal Width, Petal Length and Petal Width.
* The petal length and petal width is the most distinguishable among the group of setosa, versicolor and virginica.

### **Box Plot**

```python
sns.boxplot(data_iris, x = 'Classification', y = 'PL', hue ='Classification', palette = ['purple','green','yellow'])
plt.title('Petal Length Distributiion by Classification')
plt.show(block=True)
```

Fom the boxplot:
* Virginica has the highest and wide range of petal length.
* Setosa has the smallest petal length with a few outliers.
* Virginica and Versicolor has an overlapping range in terms of petal length thus difficult to distinguish in tyerms of petal length.

### **Bar Plot**

```python
sns.barplot(data=data_iris, x='Classification', y='SL', hue ='Classification',palette = ['purple','green','yellow'])
plt.title('Average Sepal Length by Species')
plt.show()
```

The Bar Plot depicts:
* Virginica has the largest sepal length followed by Versicolor.
* Setosa has smallest Sepal Length.
