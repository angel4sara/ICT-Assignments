**20.** Draw a plot between dep_delay and arr_delay

````python
plt.scatter(my_flights['dep_delay'], my_flights['arr_delay'],marker='*',color = 'black')
plt.title('Plot between Departure Delay and Arrival Delay')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Arrival Delay (minutes)')
plt.grid(True)
plt.show()
````

1. There is a positive trend between the Arrival Delay and Departure Delay, as Arrival Delay increases,Departure Delay also increases.
2. The clustering of the datapoints in low arrival and departure dealys indicates that most flights have mininal delays.
3. Fewer datapoints(5 datapoints) falls in the extreme delay region, thus indicating that high delay time are less frequent.
